```python
#arithmetic operations
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
print()

add=num1+num2
print("addition:",add)
print()

sub=num1-num2
print("subtraction:",sub)
print()

mul=num1*num2
print("multiplication:",mul)
print()

div=num1/num2
print("division:",div)
print()

square=num1**2
print("square of first number is:",square)
```

```python
#string
txt1 = input("Enter first string: ")
txt2 = input("Enter second string: ")
print()

concatenation = txt1 + txt2
print("Concatenation:", concatenation)
print()

print("To upper case:", txt1.upper())
print("To lower case:", txt2.lower())
print()

print("Length of first string:", len(txt1))
print("Length of second string:", len(txt2))
print()

print("Capitalize:", txt1.capitalize())
print()

print("Center:", txt1.center(6, '*'))
print()

print("Left Justify:", txt1.ljust(10, '-'))
print()

print("Right Justify:", txt2.rjust(10, '-'))
print()

print("Formating:")
print("Name1: {first}, Name2: {second}".format(first=txt1, second=txt2))
print()

old = input("Enter the word/character to replace from first string: ")
new = input("Enter the new word/character: ")
print("After Replace:", txt1.replace(old, new))
print()

print("Before:", txt2)
print("After:", txt2.strip())
```

```python
#functiion
num=int(input("Enter a number:"))

def factorial(num):
    return[i for i in range(1,num+1) if num%i==0]

print("The factorial of the number",num, " is:",factorial(num))
```

```python
# e)using function return true if 2 list have atleast one common value
def common_member(list1, list2):
    for item in list1:
        if item in list2:
            return True
    return False

list1 = [1, 2, 3, 4]
list2 = [5, 6, 3, 8]

print(common_member(list1, list2))
```
